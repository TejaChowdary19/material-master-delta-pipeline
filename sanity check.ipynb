```python
%sql
SELECT 
  'Bronze Records' as metric,
  COUNT(*) as count
FROM material_master_bronze

UNION ALL

SELECT 
  'Silver Records' as metric,
  COUNT(*) as count
FROM material_master_silver

UNION ALL

SELECT 
  'Dropped Records' as metric,
  (SELECT COUNT(*) FROM material_master_bronze) - 
  (SELECT COUNT(*) FROM material_master_silver) as count;
```

```python
%sql
SELECT 
  CASE 
    WHEN unit_cost > 0 AND unit_cost <= 500 THEN 'PASSED: Cost in range (0-500)'
    WHEN unit_cost > 500 THEN 'WARNING: Cost > 500'
    ELSE 'INVALID: Cost <= 0'
  END as expectation_status,
  COUNT(*) as record_count,
  ROUND(MIN(unit_cost), 2) as min_cost,
  ROUND(AVG(unit_cost), 2) as avg_cost,
  ROUND(MAX(unit_cost), 2) as max_cost,
  ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER(), 2) as percentage
FROM material_master_silver
GROUP BY 
  CASE 
    WHEN unit_cost > 0 AND unit_cost <= 500 THEN 'PASSED: Cost in range (0-500)'
    WHEN unit_cost > 500 THEN 'WARNING: Cost > 500'
    ELSE 'INVALID: Cost <= 0'
  END
ORDER BY record_count DESC;
```
